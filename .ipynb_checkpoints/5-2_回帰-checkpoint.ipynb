```python
# 原点を通る直線による近似　reg1dim1
def reg1dim1(x, y):
    a = np.dot(x, y) / (x**2).sum()
    return a
```

```python
x = np.array([1, 2, 4, 6, 7])
y = np.array([1, 3, 3, 5, 4])
a = reg1dim1(x, y)
print(a)
```

```python
plt.scatter(x, y, color='r')
xmax = x.max()
print(xmax)
plt.plot([0, xmax], [0, a*xmax], color='r')
plt.show()
```

```python
# 一般の直線による近似
def reg1dim2(x, y):
    n = len(x)
    a = ((np.dot(x, y) - y.sum() * x.sum() / n) /
         ((x**2).sum() - x.sum()**2 / n))
    b = (y.sum() - a * x.sum()) / n
    return a, b
```

```python
x = np.array([1, 2, 4, 6, 7])
y = np.array([1, 3, 3, 5, 4])
a, b = reg1dim2(x, y)
```

```python
print(a, b)
```

```python
plt.scatter(x, y, color='k')
xmax = x.max()
plt.plot([0, xmax], [b, a * xmax + b], color='k')
plt.show()
```

```python
# 特徴量ベクトルが多次元の場合
class LinearRegression:
    def __init__(self):
        self.w_ = None
    
    def fit(self, X, t):
        # Xtilはx0が1なので、np.ones(X.shape[0])でX行分1を追加する
        Xtil = np.c_[np.ones(X.shape[0]), X]
        A = np.dot(Xtil.T, Xtil)
        b = np.dot(Xtil.T, t)
        # linalg.solveはAx=bの形で行列計算する連立方程式を解く
        self.w_ = linalg.solve(A, b)
    
    def predict(self, X):
        if X.ndim == 1:
            X = X.reshape(1, -1)
        Xtil = np.c_[np.ones(X.shape[0]), X]
        return np.dot(Xtil, self.w_)
```

```python
import csv
```

```python
# データ読み込み
Xy = []
with open("winequality-red.csv") as fp:
    for row in csv.reader(fp, delimiter=";"):
        Xy.append(row)
Xy = np.array(Xy[1:], dtype=np.float64)
```

```python
# 訓練用データとテスト用データに分割する
np.random.seed(0)
np.random.shuffle(Xy)
train_X = Xy[:-1000, :-1]
train_y = Xy[:-1000, -1]
test_X = Xy[-1000:, :-1]
test_y = Xy[-1000:, -1]
```

```python
# 学習させる
model = LinearRegression()
model.fit(train_X, train_y)
```

```python
# テスト用データにモデルを適用
y = model.predict(test_X)
```

```python
print("最初の5つの正解と予測値：")
for i in range(5):
    print("{:1.0f} {:5.3f}".format(test_y[i], y[i]))
print()
print("RMSE:", np.sqrt(((test_y)**2).mean()))
```
